```python
from tkinter import *
from random import *
from time import *

points = 0
obratniy_otschet_c = 120
obratniy_otschet_mo = 120
obratniy_otschet_mt = 120

clr = [
    'snow','ghost white','gainsboro','linen','papaya whip','peach puff','navajo white','lemon chiffon','mint cream',
    'lavender','lavender blush','misty rose','dark slate gray','dim gray','slate gray','gray','midnight blue','navy',
    'cornflower blue','slate blue','medium blue','royal blue','dodger blue','deep sky blue','steel blue','cadet blue',
    'cyan','medium aquamarine','sea green','dark green','dark olive green','pale green','lawn green','green yellow',
    'forest green','olive drab','dark khaki','khaki','yellow','gold','goldenrod','light goldenrod','rosy brown','coral',
    'salmon','orange','dark orange','tomato','red','orange red','deep pink','hot pink','pink','maroon','dark orchid',
    'medium violet red','dark violet','purple','medium purple','thistle','snow2','snow4','seashell4','AntiqueWhite1',
    'AntiqueWhite2','AntiqueWhite3','AntiqueWhite4','ivory4','azure3','SlateBlue1','SlateBlue3','RoyalBlue2','blue4',
    'RoyalBlue4','SteelBlue4','SkyBlue2','LightSkyBlue4','cyan3','DarkSeaGreen1','DarkSeaGreen4','SeaGreen2','tan2',
    'chocolate3',
    ]

def colore():
    if points%10 == 0:
        for i in range(3):
            b_one["bg"]=choice(clr)
            b_one["fg"]=choice(clr)
            b_two["bg"]=choice(clr)
            b_two["fg"]=choice(clr)
            btn["bg"]=choice(clr)
            btn["fg"]=choice(clr)
def timer():
    global obratniy_otschet_mo
    global obratniy_otschet_mt
    global obratniy_otschet_c
    obratniy_otschet_mo -= 1
    obratniy_otschet_mt -= 1
    obratniy_otschet_c -= 1
def close_window(event=None):
    window.destroy()
def curs(event):
    global btn
    global obratniy_otschet_c
    label_points['text'] = points
    x = randrange(30, 1700)
    y = randrange(60, 1020)
    btn.place(x=x, y=y)
    colore()
    obratniy_otschet_c = 120
def move_one(event):
    global points
    global obratniy_otschet_mo
    b_one.place(x=randint(30, 1780),y=randint(60, 1020))
    points += 1
    label_points['text'] = points
    colore()
    obratniy_otschet_mo = 120
def move_two(event):
    global obratniy_otschet_mt
    obratniy_otschet_mt = 120
    global points
    b_two.place(x=randint(30, 1780),y=randint(60, 1020))
    points += 1
    label_points['text'] = points
    colore()
def lose(event):
    def close_root(event=None):
        root.destroy()
    window.destroy()
    root = Tk()
    root.title("Реакция")
    root.geometry('1920x1080')
    root.resizable(width=False,height=False)
    root.wm_attributes("-fullscreen", True)
    root.configure(bg="black")
    label_lose = Label(text='YOU LOSE', font=('Arial',45),bg="black", fg='red')
    label_lose.place(x=770, y=410)
    b_root_close = Button(text='EXIT', font=('Arial',30),bg="red", fg='black')
    b_root_close.configure(command=close_root)
    b_root_close.place(x=870, y=500)
    root.bind("<Escape>",close_root)
    root.mainloop()

window = Tk()
window.title("REACTOIN")
window.geometry('1920x1080')
window.resizable(width=False,height=False)
window.wm_attributes("-fullscreen", True)
window.configure(bg="deep sky blue")
window.bind("<Escape>",close_window)

label_points = Label(text=points, font=('Arial',20),bg="deep sky blue", fg='black')
label_points.place(x=10, y=5)
b_window_close = Button(text='CLOSE', font=('Arial',15),bg="gray15", fg='white')
b_window_close.configure(command=close_window)
b_window_close.pack(anchor=NE)
b_one = Button(text='LMB', font=('Arial', 14), fg='black')
b_one.place(x=890,y=465)
b_one.bind('<Button-1>',move_one)
b_one.bind('<Button-3>',lose)
b_two = Button(text='RMB', font=('Arial', 14), fg='white', bg='brown')
b_two.place(x=300,y=465)
btn = Button(text='ENTER', font=('Arial', 14), bg='green4', fg='white')
btn.place(x=1600,y=465)

while True:
    window.update()
    sec_mo = str(obratniy_otschet_mo//30)
    sec_mt = str(obratniy_otschet_mt//30)
    sec_c = str(obratniy_otschet_c//30)
    label_lmb = Label(text=f"LMB IS ACTIVE {sec_mo}", font=('Arial',20), bg="orange", fg="black")
    label_rmb = Label(text="RMB IS NOT ACTIVE", font=('Arial',20), bg="deep sky blue", fg="gray")
    label_enter = Label(text="ENTER IS NOT ACTIVE", font=('Arial',20), bg="deep sky blue", fg="gray")
    label_lmb.place(x=130, y=5, width=300)
    label_rmb.place(x=760, y=5, width=300)
    label_enter.place(x=1400, y=5, width=300)
    window.after(10,timer())
    if points == 21:
        obratniy_otschet_mt = 120
    if points == 31:
        obratniy_otschet_c = 120
    if points > 20:
        label_rmb.config(text=f"RMB IS ACTIVE {sec_mt}", font=('Arial',20), bg="orange", fg="black")
        b_two.config(state='normal')
        b_two.bind('<Button-3>', move_two)
        b_two.bind('<Button-1>', lose)
    else:
        b_two.config(state='disabled')
    if points > 30:
        label_enter.config(text=f"ENTER IS ACTIVE {sec_c}", font=('Arial',20), bg="orange", fg="black")
        btn.config(state='normal')
        btn.bind('<Enter>', curs)
    else:
        btn.config(state='disabled')
    if obratniy_otschet_mo <= 0:
        lose(event=1)
    if obratniy_otschet_mt <= 0 and points > 20:
        lose(event=1)
    if obratniy_otschet_c <= 0 and points > 30:
        lose(event=1)
    window.update()
    window.after(10,timer())
    if points == 21:
        obratniy_otschet_mt = 120
    if points == 31:
        obratniy_otschet_c = 120
    if points > 20:
        b_two.config(state='normal')
        b_two.bind('<Button-3>', move_two)
        b_two.bind('<Button-1>', lose)
    else:
        b_two.config(state='disabled')
    if points > 30:
        btn.config(state='normal')
        btn.bind('<Enter>', curs)
    else:
        btn.config(state='disabled')
    if obratniy_otschet_mo <= 0:
        lose(event=1)
    if obratniy_otschet_mt <= 0 and points > 20:
        lose(event=1)
    if obratniy_otschet_c <= 0 and points > 30:
        lose(event=1)
```
